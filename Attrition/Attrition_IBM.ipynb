```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
from matplotlib.pyplot import figure
```

```python
# Read data
df = pd.read_csv(r'Employee Attrition.csv')
df
```

```python
sns.set_theme(style="darkgrid")
sns.displot(
    df, x="Department",hue="Attrition",
    binwidth=7, height=7, facet_kws=dict(margin_titles=True),
)
```

```python
# Correlation matrix
plt.figure(figsize = (80,50))
sns.set(font_scale=3)
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
# Numerise all attributes
df_numerised = df
for col in df_numerised.columns:
    if (df_numerised[col].dtype == 'object'):
        df_numerised[col] = df_numerised[col].astype('category').cat.codes
```

```python
# Look at the correlation between attrition and all other factors
stacked_df_numerised = df_numerised.corr().stack().sort_values(ascending=False)
stacked_df_numerised['Attrition']
```

```python
sns.set_theme(context="notebook",style="darkgrid",palette="muted",color_codes="False")
sns.displot(
    df,kind="hist", hue="Attrition",col="MaritalStatus",x="OverTime",multiple="stack",
    binwidth=5, height=5, facet_kws=dict(margin_titles=True),
)
```
